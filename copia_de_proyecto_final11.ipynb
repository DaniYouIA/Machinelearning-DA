## Fase 5: Análisis de Clustering con DBSCAN

Vamos a utilizar el algoritmo DBSCAN (Density-Based Spatial Clustering of Applications with Noise) para identificar agrupaciones naturales en los puntajes globales (`PUNT_GLOBAL`) de los estudiantes. Este método es útil para descubrir clústeres de forma irregular y también para identificar 'ruido' (puntos que no pertenecen a ningún clúster). Antes de aplicar DBSCAN, escalaremos los datos para que el parámetro `eps` tenga un significado consistente.

````python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

print("Preparando y ejecutando DBSCAN en PUNT_GLOBAL...")

if 'df_filtrado' in locals() and not df_filtrado.empty:
    # Seleccionar la característica para clustering y eliminar valores nulos
    # Reshape para que StandardScaler espere una entrada 2D
    X = df_filtrado['PUNT_GLOBAL'].dropna().values.reshape(-1, 1)

    if X.shape[0] == 0:
        print("No hay datos válidos de PUNT_GLOBAL para el clustering.")
    else:
        # Escalar los datos
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        # Aplicar DBSCAN con los parámetros proporcionados por el usuario
        db = DBSCAN(eps=0.3, min_samples=10).fit(X_scaled)
        labels = db.labels_

        # Número de clústeres en las etiquetas, ignorando el ruido.
        n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise_ = list(labels).count(-1)

        print("\nResultados de DBSCAN:")
        print("Número estimado de clústeres: %d" % n_clusters_)
        print("Número estimado de puntos de ruido: %d" % n_noise_)

        # Añadir las etiquetas de clúster al DataFrame original (solo para los puntos no nulos usados)
        df_temp = df_filtrado['PUNT_GLOBAL'].dropna().to_frame()
        df_temp['DBSCAN_Cluster'] = labels

        # Visualización de los clústeres
        fig = plt.figure(figsize=(10, 6))
        sns.histplot(df_temp, x='PUNT_GLOBAL', hue='DBSCAN_Cluster', palette='viridis', kde=True)
        plt.title('Distribución de PUNT_GLOBAL por Clúster DBSCAN')
        plt.xlabel('Puntaje Global')
        plt.ylabel('Frecuencia')
        plt.show()
else:
    print('El DataFrame filtrado (df_filtrado) no está disponible o está vacío. Asegúrate de ejecutar las celdas anteriores.')
````

### Interpretación de los Resultados de DBSCAN

Los resultados de DBSCAN nos muestran:

*   **Número de clústeres:** La cantidad de grupos densos de estudiantes con puntajes globales similares que el algoritmo pudo identificar.
*   **Puntos de ruido:** Estudiantes con puntajes globales que no se agrupan densamente con otros, y que DBSCAN considera como *outliers* o ruido. Estos podrían representar casos excepcionales o anomalías en el rendimiento.

La visualización del histograma con los clústeres coloreados nos ayuda a entender cómo se distribuyen estos grupos a lo largo del espectro de puntajes globales.

### Paso 2: Filtrado de los últimos dos años

### Entendiendo los Estratos Socioeconómicos en Colombia

En Colombia, la estratificación socioeconómica es una clasificación de las viviendas que el Estado utiliza para asignar subsidios y cobrar tarifas diferenciales por los servicios públicos domiciliarios. Los estratos van del 1 al 6, siendo:

*   **Estratos 1 y 2 (Bajo-Bajo y Bajo):** Hogares con menores recursos, con derecho a subsidios en servicios públicos.
*   **Estratos 3 y 4 (Medio-Bajo y Medio):** Hogares con recursos intermedios.
*   **Estratos 5 y 6 (Medio-Alto y Alto):** Hogares con mayores recursos, que deben pagar un sobrecosto (contribución) en sus servicios públicos.

El estrato `Sin Estrato` generalmente agrupa a estudiantes cuya vivienda no fue clasificada o para quienes la información no está disponible. Esta variable es clave en nuestro análisis para entender la desigualdad en el rendimiento académico.
